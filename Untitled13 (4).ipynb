```python
from pypdf import PdfReader

reader = PdfReader("EJ1172284.pdf")

text = ""

for page in reader.pages:
    content = page.extract_text()
    if content:
        text += content

print("Total text length:", len(text))
print(text[:1000])
```

```python
from langchain_text_splitters import CharacterTextSplitter

splitter = CharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = splitter.split_text(text)

print("Number of chunks:", len(chunks))
print("\nExample chunk:\n")
print(chunks[0])
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = model.encode(chunks)

print("Embedding shape:", len(embeddings))
```

```python
query = input("Ask a question about the paper: ")
```

```python
query_embedding = model.encode([query])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

scores = cosine_similarity(query_embedding, embeddings)[0]

top_indices = np.argsort(scores)[-3:][::-1]

print("\nMost relevant sections:\n")

for idx in top_indices:
    print(chunks[idx])
    print("\n---------------------\n")
```

```python
context = " ".join([chunks[i] for i in top_indices])

print("Context used for answering:\n")
print(context[:1000])
```

```python
while True:

    query = input("\nAsk a question (type 'exit' to stop): ")

    if query.lower() == "exit":
        break

    query_embedding = model.encode([query])

    scores = cosine_similarity(query_embedding, embeddings)[0]

    top_indices = np.argsort(scores)[-3:][::-1]

    context = " ".join([chunks[i] for i in top_indices])

    print("\nAnswer based on paper:\n")
    print(context[:800])
```
